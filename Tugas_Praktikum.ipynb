```python
from tabulate import tabulate

# ====== Data Global ======
dimsum_stock = {}
riwayat_pembelian = []  # List of dict
total_pemasukan = 0

# ====== Fungsi Utility ======
def input_angka(teks, minimal=1):
    try:
        angka = int(input(teks))
        if angka < minimal:
            print(f"Masukkan angka lebih dari {minimal - 1}.")
            return None
        return angka
    except ValueError:
        print("Masukkan angka yang valid.")
        return None

def tampilkan_stok():
    if not dimsum_stock:
        print("Belum ada dimsum yang tersedia.")
        return
    tabel = [[nama, info['jumlah'], f"Rp {info['harga']:,}"] for nama, info in dimsum_stock.items()]
    print(tabulate(tabel, headers=["Nama Dimsum", "Jumlah", "Harga"], tablefmt="grid"))

# ====== Menu Penjual ======
def menu_penjual():
    global total_pemasukan
    while True:
        print("\nüë®‚Äçüíº MENU PENJUAL")
        print("1. Lihat pemasukan")
        print("2. Tambah stok dimsum")
        print("3. Lihat stok dimsum")
        print("4. Lihat riwayat pembelian")
        print("5. Kembali")

        pilihan = input("Pilih menu: ")
        if pilihan == "1":
            print(f"üí∞ Total pemasukan: Rp {total_pemasukan:,}")
        elif pilihan == "2":
            nama = input("Nama dimsum: ").title()
            jumlah = input_angka("Jumlah dimsum: ")
            if jumlah is None:
                continue
            harga = input_angka("Harga per pcs: ")
            if harga is None:
                continue

            if nama in dimsum_stock:
                dimsum_stock[nama]['jumlah'] += jumlah
                dimsum_stock[nama]['harga'] = harga
            else:
                dimsum_stock[nama] = {'jumlah': jumlah, 'harga': harga}
            print(f"‚úÖ Dimsum '{nama}' berhasil ditambahkan/diperbarui.")
        elif pilihan == "3":
            tampilkan_stok()
        elif pilihan == "4":
            if not riwayat_pembelian:
                print("Belum ada pembelian.")
            else:
                print("\nüßæ Riwayat Pembelian:")
                for data in riwayat_pembelian:
                    pesanan_str = ', '.join([f"{nama}({qty})" for nama, qty in data['pesanan'].items()])
                    print(f"- {data['nama']} beli {pesanan_str} | Total: Rp {data['total']:,}")
        elif pilihan == "5":
            break
        else:
            print("‚ùå Pilihan tidak valid.")

# ====== Menu Pembeli ======
def menu_pembeli():
    global total_pemasukan
    while True:
        print("\nüßç MENU PEMBELI")
        print("1. Lihat stok dimsum")
        print("2. Pesan dimsum")
        print("3. Kembali")

        pilihan = input("Pilih menu: ")
        if pilihan == "1":
            tampilkan_stok()
        elif pilihan == "2":
            if not dimsum_stock:
                print("Dimsum belum tersedia.")
                continue

            nama_pembeli = input("Masukkan nama kamu: ").title()
            pesanan = {}
            total_harga = 0

            while True:
                tampilkan_stok()
                nama = input("Pilih nama dimsum (atau 'selesai'): ").title()
                if nama.lower() == "selesai":
                    break
                if nama not in dimsum_stock:
                    print("Dimsum tidak ditemukan.")
                    continue

                jumlah = input_angka(f"Jumlah {nama}: ")
                if jumlah is None:
                    continue

                stok = dimsum_stock[nama]['jumlah']
                harga = dimsum_stock[nama]['harga']
                if jumlah > stok:
                    print("Stok tidak mencukupi.")
                    continue

                pesanan[nama] = pesanan.get(nama, 0) + jumlah
                dimsum_stock[nama]['jumlah'] -= jumlah
                total_harga += jumlah * harga

            if pesanan:
                print(f"\nüßæ Total yang harus dibayar: Rp {total_harga:,}")
                riwayat_pembelian.append({
                    "nama": nama_pembeli,
                    "pesanan": pesanan,
                    "total": total_harga
                })
                total_pemasukan += total_harga
            else:
                print("‚ùå Tidak ada pesanan yang dilakukan.")
        elif pilihan == "3":
            break
        else:
            print("‚ùå Pilihan tidak valid.")

# ====== Main Program ======
def main():
    while True:
        print("\n===== SELAMAT DATANG DI TOKO DIMSUM D'LICIOUS =====")
        print("1. Penjual")
        print("2. Pembeli")
        print("3. Keluar")

        pilihan = input("Pilih peran: ")
        if pilihan == "1":
            menu_penjual()
        elif pilihan == "2":
            menu_pembeli()
        elif pilihan == "3":
            print("üëã Hanupis, Hatur Nuhun Pisan, Euy!")
            break
        else:
            print("‚ùå Pilihan tidak valid.")

# Jalankan program
main()
```
